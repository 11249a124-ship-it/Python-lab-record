Q6="Similarity: Write a Python program to find the string similarity between two given strings."
print(Q6)

```python
from difflib import SequenceMatcher
def string_similarity(str1, str2):
     matcher = SequenceMatcher(None, str1, str2)
     return matcher.ratio()
s1 = "Python Exercises"
s2 = "Python Exercise"
print("Similarity:", string_similarity(s1, s2))
```
